```python
import numpy as np
import pandas as pd
import seaborn as sns

from scipy.stats import f_oneway
from matplotlib import pyplot as plt

sns.set_style('whitegrid')
sns.set_context('poster')

%matplotlib inline
```

```python
data = dict(New = [50,39,42,45,38,44,40,49,42,41],
            Old = [44,31,50,22,30,27,32,25,40],
            Control = [16,60,24,19,31,37,44,55],
           )
n_permutations = int(1e4)
f_original,p = f_oneway(data['New'],
                        data['Old'],
                        data['Control'])
```

```python
def sampling(data:dict) -> np.ndarray:
    """
    Sample data from the original dataset with replacement
    
    Input
    ---
    data:dictionary
    
    Output
    ---
    samples:numpy.ndarray (n_samples,)
    """
    pooled_data = np.concatenate([item for key,item in data.items()])
    samples = np.random.choice(pooled_data,size = pooled_data.shape,replace = True,)
    return samples
def compute_anova_on_sampled_data(data:dict) -> float:
    """
    1. sample data from the original data
    2. reshape the sampled data
    3. compute the F value
    
    Input
    ---
    data:dictionary
    
    Output
    ---
    F : float, the F statistics calculated by f_oneway
    
    """
    samples = sampling(data)
    a = samples[:10]
    b = samples[10:10+9]
    c = samples[-8:]
    f,p = f_oneway(a,b,c)
    return f
```

```python
np.random.seed(12345)
Fs = np.array([compute_anova_on_sampled_data(data) for _ in range(n_permutations)])
```

```python
fig,axes = plt.subplots(figsize = (12,6),
                        nrows = 2,
                        sharex = True,)
ax = sns.kdeplot(Fs,cut = 0,ax = axes[0])
ax = sns.histplot(Fs,bins = 20,ax = axes[1])
ax.set(xlim = (-0.5,7))
plt.tight_layout()
```

```python
pval = (np.sum(Fs >= f_original) + 1) / (n_permutations + 1)
print(f'original F(2,15) = {f_original:.4f}, sampled F(2,15) = {np.mean(Fs):.4f},p = {pval:.4f}')
```
